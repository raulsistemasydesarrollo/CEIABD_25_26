# IntroducciÃ³n a **Polars** âš¡

## Â¿QuÃ© es Polars?

**Polars** es una **biblioteca de Python** increÃ­blemente **rÃ¡pida** diseÃ±ada para trabajar con **DataFrames** (conjuntos de datos tabulares).
EstÃ¡ escrita en **Rust** ğŸ¦€ y utiliza **Apache Arrow** como formato de memoria, lo que le permite procesar grandes volÃºmenes de datos de forma muy eficiente, a menudo mucho mÃ¡s rÃ¡pido que Pandas.

Permite **analizar, limpiar, transformar y consultar datos** con una sintaxis expresiva y optimizada para el rendimiento.

Fue creada por **Ritchie Vink** y estÃ¡ ganando popularidad rÃ¡pidamente en la comunidad de ciencia de datos y *Machine Learning* por su velocidad y eficiencia en el uso de memoria.

---

## Â¿Por quÃ© usar Polars?

En proyectos de **Inteligencia Artificial** y **Machine Learning**, especialmente con **Big Data**, la velocidad y la gestiÃ³n eficiente de la memoria son cruciales.

Polars destaca porque:

* **Es rapidÃ­simo**: Aprovecha todos los nÃºcleos de tu CPU para paralelizar operaciones.
* **Maneja datos grandes**: Puede procesar datasets que no caben completamente en la RAM gracias a su motor *lazy* y optimizaciones.
* **Tiene una API expresiva**: Permite encadenar operaciones de forma clara y lÃ³gica (similar a SQL o dplyr en R).
* **Integra Apache Arrow**: Facilita la interoperabilidad con otras herramientas del ecosistema de datos.

 Polars te ayuda a realizar las mismas tareas que Pandas (limpiar, explorar, transformar), pero a menudo con un **rendimiento significativamente mayor**.

---

## ConexiÃ³n con la Ciencia de Datos

Polars es una herramienta moderna y potente dentro de la **Ciencia de Datos**.
Se posiciona como una alternativa (o complemento) a Pandas, especialmente Ãºtil cuando el **rendimiento** es un factor crÃ­tico o se trabaja con **datasets muy grandes**.

Se integra bien con otras bibliotecas como **NumPy**, **PyArrow**, y herramientas de visualizaciÃ³n como **Matplotlib** o **Seaborn** (a menudo convirtiendo los datos a Pandas o NumPy para graficar).

---

## Â¿QuÃ© puede hacer Polars?

Con Polars puedes:

* Leer y escribir datos desde/hacia mÃºltiples formatos (CSV, Parquet, JSON, bases de datos...).
* Filtrar, ordenar, agrupar y agregar datos de forma muy eficiente.
* Unir (`join`) y concatenar (`concat`) diferentes DataFrames.
* Crear nuevas columnas basadas en expresiones complejas.
* Manejar valores nulos (`null`).
* Trabajar con datos de series temporales, cadenas de texto, y mÃ¡s.

Una de sus caracterÃ­sticas clave es su **motor de ejecuciÃ³n *lazy*** (perezoso), que optimiza las consultas antes de ejecutarlas.

# **Polars Series** ğŸ“Š

## Â¿QuÃ© es una *Series* en Polars?

Una **Series** de Polars representa una **Ãºnica columna de datos** en un DataFrame.
Es una estructura **unidimensional** optimizada para el rendimiento que contiene datos de un **Ãºnico tipo** (nÃºmeros, texto, fechas, etc.).

Puedes pensar en ella como una columna de una tabla de base de datos o una lista/array de NumPy con funcionalidades adicionales.

**Diferencia clave con Pandas:** Las Series en Polars **no tienen un Ã­ndice explÃ­cito** como en Pandas. El acceso se basa en la posiciÃ³n o mediante expresiones de filtrado.

---

## Crear una *Series*

Puedes crear una Series a partir de diferentes estructuras de datos de Python, como listas o tuplas.

ğŸ“˜ **Ejemplo 1: Desde una lista**

````python
import polars as pl

datos_lista = [10, 20, 15, 30, 20]
serie_numeros = pl.Series("valores", datos_lista)

print(serie_numeros)
````

**Nota:** Al crear una Series, es obligatorio darle un **nombre** (el primer argumento, "valores" en este caso).

ğŸ“˜ **Ejemplo 2: Serie de texto**

````python
datos_texto = ["manzana", "plÃ¡tano", "manzana", "naranja"]
serie_frutas = pl.Series("frutas", datos_texto)

print(serie_frutas)
````

---

## Acceder a elementos de una *Series*

Puedes acceder a los elementos por su **posiciÃ³n** (Ã­ndice basado en 0), similar a las listas de Python.

ğŸ“˜ **Ejemplo: Acceder al primer y tercer elemento**

````python
# Acceder al primer elemento (posiciÃ³n 0)
print(f"Primer valor: {serie_numeros[0]}")

# Acceder al tercer elemento (posiciÃ³n 2)
print(f"Tercer valor: {serie_numeros[2]}")
````

TambiÃ©n puedes usar **slicing** para obtener un subconjunto de la Series:

````python
# Obtener los elementos desde la posiciÃ³n 1 hasta la 3 (sin incluir la 4)
print("Slice [1:4]:")
print(serie_numeros[1:4])
````

---

## Operaciones bÃ¡sicas con *Series*

Las Series de Polars permiten realizar operaciones vectorizadas (aplicadas a todos los elementos a la vez), lo cual es muy eficiente.

ğŸ“˜ **Ejemplo: Operaciones aritmÃ©ticas y comparaciones**

````python
# Sumar 5 a cada elemento
print("Serie + 5:")
print(serie_numeros + 5)

# Multiplicar por 2
print("\nSerie * 2:")
print(serie_numeros * 2)

# Comparar si los elementos son mayores que 15
print("\nSerie > 15 (devuelve booleanos):")
print(serie_numeros > 15)
````

TambiÃ©n puedes aplicar funciones estadÃ­sticas bÃ¡sicas:


````python
print(f"\nSuma total: {serie_numeros.sum()}")
print(f"Valor medio: {serie_numeros.mean()}")
print(f"Valor mÃ¡ximo: {serie_numeros.max()}")
print(f"NÃºmero de elementos Ãºnicos: {serie_frutas.n_unique()}")
print("Valores Ãºnicos y sus cuentas:")
print(serie_frutas.value_counts())
````

## De Series a DataFrame

Un **DataFrame** en Polars es esencialmente una **colecciÃ³n de Series** con el mismo nÃºmero de filas, donde cada Series representa una columna.

Puedes construir un DataFrame a partir de varias Series:

ğŸ“˜ **Ejemplo:**

````python
import polars as pl

serie_calorias = pl.Series("calories", [420, 380, 390])
serie_duracion = pl.Series("duration", [50, 40, 45])

# Crear DataFrame a partir de las Series
df_entrenamiento = pl.DataFrame([serie_calorias, serie_duracion])

print(df_entrenamiento)
````

Cada Series se convierte en una columna del DataFrame, usando el nombre de la Series como cabecera de la columna.

# IntroducciÃ³n a los DataFrames en Polars ğŸš€

## Â¿QuÃ© es un DataFrame en Polars?

Un **DataFrame** de **Polars** es la estructura de datos principal para trabajar con **datos tabulares** (organizados en filas y columnas).
Es similar a una tabla de base de datos, una hoja de cÃ¡lculo de Excel, o un DataFrame de Pandas, pero estÃ¡ diseÃ±ado desde cero para ser **extremadamente rÃ¡pido y eficiente con la memoria**.

ğŸ”¹ Es **bidimensional** (filas y columnas).
ğŸ”¹ Cada columna es una **Polars Series** con un nombre y un tipo de dato Ãºnico.
ğŸ”¹ **No tiene un Ã­ndice** como Pandas; las filas se identifican por su posiciÃ³n.
ğŸ”¹ Optimizado para **procesamiento en paralelo** y **manejo de grandes datasets**.

---

## Crear un DataFrame

Puedes crear un DataFrame de Polars a partir de varias fuentes:

ğŸ“˜ **Ejemplo 1: Desde un diccionario de listas (como en Pandas)**

````python
import polars as pl

data_dict = {
  "columna_a": [1, 2, 3, 4, 5],
  "columna_b": ["A", "B", "C", "D", "E"],
  "columna_c": [True, False, True, False, True]
}

df_dict = pl.DataFrame(data_dict)
print(df_dict)
````


ğŸ“˜ **Ejemplo 2: Desde una lista de diccionarios (cada dict es una fila)**

Esto es Ãºtil si tus datos vienen en formato JSON de registros.

````python
data_list_dict = [
    {"id": 1, "producto": "Laptop", "precio": 1200},
    {"id": 2, "producto": "Teclado", "precio": 75},
    {"id": 3, "producto": "Mouse", "precio": 25}
]

df_list_dict = pl.DataFrame(data_list_dict)
print(df_list_dict)
````


ğŸ“˜ **Ejemplo 3: Desde un array de NumPy**

Necesitas proporcionar los nombres de las columnas.

````python
import numpy as np

data_np = np.array([[1.0, 2.5], [3.0, 4.5], [5.0, 6.5]])
column_names = ["sensor_1", "sensor_2"]

df_np = pl.DataFrame(data_np, schema=column_names)
print(df_np)
````

## Seleccionar Filas y Columnas

Polars utiliza una sintaxis basada en **expresiones** para seleccionar y manipular datos, lo que es muy potente.

### Seleccionar Columnas (`.select()`)

Usa el mÃ©todo `.select()` para elegir quÃ© columnas mostrar.

````python
# Seleccionar solo 'columna_a' y 'columna_c'
seleccion_cols = df_dict.select([
    pl.col("columna_a"),
    pl.col("columna_c")
])
print(seleccion_cols)
````

**Nota:** `pl.col("nombre")` crea una **expresiÃ³n** que se refiere a esa columna. Polars trabaja mucho con expresiones.

---

### Seleccionar Filas (`.filter()`)

Usa `.filter()` con una expresiÃ³n de condiciÃ³n para seleccionar filas.

````python
# Seleccionar filas donde 'columna_a' sea mayor que 3
filtro_filas = df_dict.filter(
    pl.col("columna_a") > 3
)
print(filtro_filas)

# Seleccionar filas donde 'columna_c' sea True
filtro_boolean = df_dict.filter(
    pl.col("columna_c") == True # O simplemente pl.col("columna_c")
)
print("\nFilas con columna_c == True:")
print(filtro_boolean)
````


---

### Acceder por posiciÃ³n (slicing)

Puedes usar slicing similar a Python/NumPy para seleccionar un rango de filas.

````python
# Seleccionar las primeras 3 filas
print("Primeras 3 filas:")
print(df_dict[0:3])

# Seleccionar filas desde la posiciÃ³n 2 hasta el final
print("\nDesde la fila 2:")
print(df_dict[2:])
````

# **LecciÃ³n: Lazy Evaluation en Polars** â³â¡ï¸ğŸš€

---

## 1. Â¿QuÃ© es la EjecuciÃ³n *Lazy* (Perezosa)?

Una de las caracterÃ­sticas mÃ¡s importantes y potentes de Polars es su **motor de ejecuciÃ³n *lazy***.
A diferencia de Pandas, que ejecuta cada operaciÃ³n inmediatamente (*eager execution*), Polars **no ejecuta las operaciones hasta que es estrictamente necesario**.

**Â¿CÃ³mo funciona?**

1.  **Plan de Consulta**: Cuando escribes una cadena de operaciones (`.select()`, `.filter()`, `.group_by()`, etc.) sobre un `LazyFrame`, Polars no calcula nada todavÃ­a. Simplemente **construye un plan lÃ³gico** de lo que quieres hacer.
2.  **OptimizaciÃ³n**: Antes de ejecutar, Polars **analiza y optimiza** este plan. Puede reordenar operaciones, fusionarlas, o elegir algoritmos mÃ¡s eficientes segÃºn el plan completo.
3.  **EjecuciÃ³n**: Solo cuando pides explÃ­citamente el resultado (normalmente con el mÃ©todo `.collect()`), Polars ejecuta el plan optimizado, a menudo en paralelo, para obtener el DataFrame final.

**AnalogÃ­a:** Imagina que pides a un chef que prepare varios platos. En lugar de cocinar cada uno por separado (eager), el chef anota todo el pedido (plan), piensa la mejor forma de usar los hornos y fuegos a la vez (optimizaciÃ³n), y solo entonces empieza a cocinar todo eficientemente (ejecuciÃ³n con `.collect()`).

---

## 2. Beneficios de la EjecuciÃ³n *Lazy*

* **Mayor Rendimiento**: La optimizaciÃ³n del plan permite a Polars encontrar la forma mÃ¡s rÃ¡pida de obtener el resultado, evitando cÃ¡lculos intermedios innecesarios.
* **Menor Uso de Memoria**: No se crean DataFrames intermedios para cada paso, reduciendo el consumo de RAM.
* **Procesamiento Out-of-Core**: Permite trabajar con datasets mÃ¡s grandes que la memoria disponible, ya que Polars puede procesar los datos en *chunks* (trozos) si es necesario (especialmente con formatos como Parquet).
* **Claridad del CÃ³digo**: Fomenta escribir cadenas de operaciones lÃ³gicas que describen la transformaciÃ³n completa.

---

## 3. Â¿CÃ³mo usar el Modo *Lazy*?

Para trabajar en modo *lazy*, necesitas empezar con un `LazyFrame` en lugar de un `DataFrame`.

### Creando un `LazyFrame`

ğŸ”¹ **Desde un archivo (mÃ©todo recomendado)**: Usa `pl.scan_csv()`, `pl.scan_parquet()`, `pl.scan_json()`, etc. Estas funciones *escanean* el archivo para conocer su estructura, pero no cargan los datos inmediatamente.


````python
import polars as pl

url_pokemon = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Pokemon.csv"

# Escanear el CSV para crear un LazyFrame
lazy_df = pl.scan_csv(url_pokemon)

print(type(lazy_df)) # VerÃ¡s que es un LazyFrame
````

ğŸ”¹ **Desde un DataFrame existente**: Usa el mÃ©todo `.lazy()`.

````python
# Crear un DataFrame eager primero (menos comÃºn para empezar)
eager_df = pl.DataFrame({"a": [1, 2], "b": [3, 4]})

# Convertirlo a LazyFrame
lazy_from_eager = eager_df.lazy()

print(type(lazy_from_eager))
````

### Aplicando Operaciones

Puedes aplicar las mismas operaciones (`.select`, `.filter`, `.with_columns`, `.group_by`, etc.) a un `LazyFrame`. Polars solo aÃ±adirÃ¡ pasos al plan de consulta.

````python
# Construir un plan sobre el LazyFrame de PokÃ©mon
plan_consulta = (
    lazy_df
    .filter(pl.col("Type 1") == "Water") # Filtrar PokÃ©mon de agua
    .filter(pl.col("Attack") > 100)      # Con ataque mayor a 100
    .select(["Name", "Attack", "Defense"]) # Seleccionar columnas
    .sort("Attack", descending=True)      # Ordenar por ataque
)

print("Tipo del plan:", type(plan_consulta)) # Sigue siendo un LazyFrame
print("\nPlan de consulta (simplificado):")
print(plan_consulta.explain(optimized=True)) # Muestra el plan optimizado
````

**Nota:** `.explain()` es Ãºtil para ver cÃ³mo Polars optimiza tu consulta.

### Ejecutando la Consulta (`.collect()`)

Para obtener el resultado final como un `DataFrame` (eager), usa `.collect()`.

````python
# Ejecutar el plan y obtener el DataFrame resultante
resultado_df = plan_consulta.collect()

print("\nTipo del resultado:", type(resultado_df))
print("\nResultado final (PokÃ©mon de Agua con Ataque > 100):")
print(resultado_df)
````

**Solo en este Ãºltimo paso (`.collect()`) Polars realmente lee el archivo CSV, aplica los filtros y selecciones optimizados, y devuelve el resultado.**

---

## 4. Â¿CuÃ¡ndo usar *Lazy* vs *Eager*?

* **Usa *Lazy* (`scan_...`, `.lazy()`, `.collect()`):**
    * Para **datasets grandes** o que no caben en memoria.
    * Cuando realizas **mÃºltiples operaciones encadenadas** (filtros, selecciones, agregaciones).
    * Para obtener el **mÃ¡ximo rendimiento**.
    * **Casi siempre es la forma recomendada** de empezar en Polars, especialmente leyendo archivos.

* **Usa *Eager* (`pl.DataFrame()`, `pl.read_...`):**
    * Para **datasets pequeÃ±os** donde la optimizaciÃ³n no es crÃ­tica.
    * Cuando necesitas **resultados intermedios inmediatos** (menos comÃºn en Polars).
    * Al crear DataFrames pequeÃ±os **directamente en memoria** (ej. desde diccionarios).

---

## âœ… **Resumen**

| Concepto            | DescripciÃ³n                                                                          |
| ------------------- | ------------------------------------------------------------------------------------ |
| **Lazy Evaluation** | Polars planifica y optimiza las operaciones antes de ejecutarlas.                    |
| **LazyFrame** | Objeto que representa un plan de consulta, no los datos en sÃ­.                       |
| **Crear LazyFrame** | `pl.scan_csv()`, `pl.scan_parquet()`, `df.lazy()`                                     |
| **Construir Plan** | Aplicar `.select()`, `.filter()`, etc., a un `LazyFrame`.                            |
| **Ejecutar Plan** | `.collect()`: Ejecuta el plan optimizado y devuelve un `DataFrame` (eager).          |
| **Ver Plan** | `.explain()`: Muestra el plan de ejecuciÃ³n lÃ³gico u optimizado.                      |
| **Beneficios** | Mejor rendimiento, menor uso de memoria, capacidad para datos grandes (out-of-core). |

# **LecciÃ³n: La API de Expresiones de Polars** âœ¨

---

## 1. Â¿QuÃ© son las Expresiones en Polars?

Las **expresiones** son el corazÃ³n â¤ï¸ de la API de Polars. Son la forma en que **describes las operaciones** que quieres realizar sobre tus columnas, sin decirle a Polars *cÃ³mo* hacerlas exactamente.

Piensa en una expresiÃ³n como una **receta** o una **fÃ³rmula** que Polars puede entender, optimizar y aplicar a los datos.

**CaracterÃ­sticas Clave:**

* **Declarativas**: Describen *quÃ©* quieres hacer, no *cÃ³mo*.
* **Componibles**: Puedes combinar expresiones simples para crear transformaciones complejas.
* **Optimizables**: El motor *lazy* de Polars las optimiza antes de la ejecuciÃ³n.
* **Paralelizables**: Polars puede ejecutar expresiones en paralelo sobre diferentes partes de los datos.

La funciÃ³n principal para crear expresiones es `pl.col("nombre_columna")`, que selecciona una columna.

---

## 2. Contextos de Uso de Expresiones

Las expresiones se utilizan dentro de varios mÃ©todos clave de los DataFrames (o LazyFrames):

* **`.select()`**: Para seleccionar columnas existentes o crear nuevas.
* **`.filter()`**: Para seleccionar filas basadas en condiciones.
* **`.with_columns()`**: Para aÃ±adir o modificar columnas.
* **`.group_by().agg()`**: Para realizar agregaciones (sumas, medias, etc.) sobre grupos.
* **`.sort()`**: Para ordenar el DataFrame.

---

## 3. Ejemplos de Expresiones Comunes

Vamos a usar un DataFrame de ejemplo:

````python
import polars as pl

df = pl.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "categoria": ["A", "B", "A", "B", "C"],
    "valor": [10, 25, 10, 30, 15],
    "coste": [5, 10, 8, 12, 7]
})
print(df)
````

### ğŸ”¹ SelecciÃ³n Simple (`pl.col()`)

````python
# Seleccionar las columnas 'id' y 'valor'
df_seleccion = df.select([
    pl.col("id"),
    pl.col("valor")
])
print(df_seleccion)
````

### ğŸ”¹ Alias (`.alias()`)

Para renombrar una columna en la salida.

````python
# Seleccionar 'valor' y renombrarla a 'precio'
df_alias = df.select([
    pl.col("valor").alias("precio")
])
print(df_alias)
````

### ğŸ”¹ Operaciones AritmÃ©ticas

Puedes realizar cÃ¡lculos directamente en las expresiones.

````python
# Calcular el beneficio (valor - coste)
df_beneficio = df.with_columns([
    (pl.col("valor") - pl.col("coste")).alias("beneficio")
])
print(df_beneficio)
````

### ğŸ”¹ Comparaciones y LÃ³gica Booleana (`==`, `>`, `&`, `|`)

Se usan principalmente en `.filter()`.

````python
# Filtrar filas donde categoria es 'A' Y valor es 10
df_filtrado = df.filter(
    (pl.col("categoria") == "A") & (pl.col("valor") == 10)
)
print(df_filtrado)
````

### ğŸ”¹ Funciones de AgregaciÃ³n (`.sum()`, `.mean()`, `.count()`, etc.)

Se usan tÃ­picamente dentro de `.agg()` despuÃ©s de un `.group_by()`.

````python
# Calcular la suma y la media de 'valor' por 'categoria'
df_agregado = df.group_by("categoria").agg([
    pl.sum("valor").alias("suma_valor"),
    pl.mean("valor").alias("media_valor")
])
print(df_agregado)
````

### ğŸ”¹ Funciones sobre Cadenas (`.str`)

Polars tiene un potente namespace `.str` para manipular texto.

````python
# Convertir 'categoria' a minÃºsculas
df_minusculas = df.with_columns([
    pl.col("categoria").str.to_lowercase().alias("categoria_minuscula")
])
print(df_minusculas)
````

### ğŸ”¹ Expresiones Condicionales (`pl.when().then().otherwise()`)

Similar a un `IF-ELSE` en SQL o Excel.

````python
# Crear columna 'tipo_valor' basado en 'valor'
df_condicional = df.with_columns([
    pl.when(pl.col("valor") > 20)
      .then(pl.lit("Alto")) # pl.lit() crea un valor literal
      .when(pl.col("valor") > 10)
      .then(pl.lit("Medio"))
      .otherwise(pl.lit("Bajo"))
      .alias("tipo_valor")
])
print(df_condicional)
````

---

## 4. El Poder del Encadenamiento

La verdadera potencia viene de **encadenar** estas expresiones y mÃ©todos para construir transformaciones complejas de forma legible.

````python
# Ejemplo encadenado: Filtrar, crear columna, agrupar y ordenar
resultado_complejo = (
    df
    .filter(pl.col("coste") < 10) # 1. Filtrar por coste
    .with_columns([ # 2. Calcular beneficio
        (pl.col("valor") - pl.col("coste")).alias("beneficio")
    ])
    .group_by("categoria") # 3. Agrupar por categoria
    .agg([ # 4. Calcular beneficio medio y contar
        pl.mean("beneficio").alias("beneficio_medio"),
        pl.count().alias("num_items") # pl.count() cuenta filas por grupo
    ])
    .sort("beneficio_medio", descending=True) # 5. Ordenar
)

print(resultado_complejo)
````


Este estilo es muy comÃºn en Polars y es clave para aprovechar el motor *lazy* (si empezaras con `pl.scan_csv`, todo esto serÃ­a parte del plan optimizado).

---

## âœ… **Resumen**

| CaracterÃ­stica              | DescripciÃ³n                                                               |
| --------------------------- | ------------------------------------------------------------------------- |
| **ExpresiÃ³n** | FÃ³rmula/receta que describe una operaciÃ³n sobre columnas.                 |
| **CreaciÃ³n** | `pl.col("nombre")`, operaciones (+, -, *, /), funciones (`.sum()`, etc.) |
| **Contextos Principales** | `.select()`, `.filter()`, `.with_columns()`, `.agg()`                     |
| **Alias** | `.alias("nuevo_nombre")` para renombrar el resultado de una expresiÃ³n.    |
| **Condicionales** | `pl.when().then().otherwise()` para lÃ³gica if-else.                       |
| **Literal** | `pl.lit(valor)` para usar un valor constante en una expresiÃ³n.            |
| **Encadenamiento (Chaining)** | Conectar mÃºltiples mÃ©todos para construir transformaciones complejas.     |
| **Beneficio** | CÃ³digo legible, optimizable y paralelizable por el motor de Polars.       |

# **LecciÃ³n: Limpieza de Datos con Polars** ğŸ§¹

---

## 1. Importancia de la Limpieza de Datos

Al igual que con Pandas, la **limpieza de datos** es un paso crucial antes de cualquier anÃ¡lisis o modelado en Polars. Los datos del mundo real rara vez son perfectos y suelen contener:

* **Valores Nulos (`null`)**: Datos faltantes.
* **Formatos Incorrectos**: Ej. nÃºmeros como texto, fechas inconsistentes.
* **Datos ErrÃ³neos**: Valores imposibles o fuera de rango (ej. edad negativa).
* **Duplicados**: Filas idÃ©nticas repetidas.

Polars ofrece herramientas eficientes para abordar estos problemas, aprovechando su API de expresiones y su motor *lazy*.

---

## 2. Manejo de Valores Nulos (`null`)

Polars usa `null` para representar datos faltantes.

ğŸ“˜ **Dataset de ejemplo con nulos:**

````python
import polars as pl
import numpy as np # Lo usaremos para introducir NaNs fÃ¡cilmente

df_nulos = pl.DataFrame({
    "col1": [1, 2, None, 4, 5],
    "col2": ["a", None, "c", "d", "e"],
    "col3": [10.0, 20.0, 30.0, None, 50.0]
})
print(df_nulos)
````

### ğŸ”¹ Contar Nulos
Usa `.null_count()`.

````python
print(df_nulos.null_count())
````

### ğŸ”¹ Eliminar Filas con Nulos (`.drop_nulls()`)
Elimina todas las filas que contengan al menos un valor nulo.

````python
df_sin_nulos = df_nulos.drop_nulls()
print(df_sin_nulos)
````

Puedes especificar columnas concretas con el argumento `subset`: `df_nulos.drop_nulls(subset=["col1", "col3"])`

### ğŸ”¹ Rellenar Nulos (`.fill_null()`)
Reemplaza los valores nulos con un valor especÃ­fico o una estrategia (como la media, mediana, etc.).

````python
# Rellenar con un valor fijo (ej. 0 para col1, 'desconocido' para col2)
df_relleno_fijo = df_nulos.with_columns([
    pl.col("col1").fill_null(0),
    pl.col("col2").fill_null("desconocido")
])
print("Relleno con valor fijo:")
print(df_relleno_fijo)

# Rellenar col3 con la media de esa columna
df_relleno_media = df_nulos.with_columns([
    pl.col("col3").fill_null(pl.mean("col3")) # ExpresiÃ³n dentro de fill_null!
])
print("\nRelleno con la media:")
print(df_relleno_media)
````

**Nota:** `.fill_null()` puede aceptar un valor literal o *otra expresiÃ³n* de Polars (como `pl.mean()`, `pl.median()`, `pl.lit()`), lo que lo hace muy flexible.

---

## 3. Corregir Formatos Incorrectos (`.cast()`)

Si una columna tiene el tipo de dato incorrecto (ej. un nÃºmero almacenado como texto), usa el mÃ©todo `.cast()` para convertirla.

ğŸ“˜ **Ejemplo:**

````python
df_formatos = pl.DataFrame({
    "id": ["1", "2", "3"], # NÃºmeros como texto
    "fecha_texto": ["2023-01-01", "2023-01-02", "2023-01-03"]
})
print("DataFrame Original:")
print(df_formatos)
print(df_formatos.dtypes)

# Convertir 'id' a entero y 'fecha_texto' a fecha
df_corregido = df_formatos.with_columns([
    pl.col("id").cast(pl.Int64), # Convertir a entero de 64 bits
    pl.col("fecha_texto").str.strptime(pl.Date, "%Y-%m-%d").alias("fecha") # Convertir texto a fecha
]).drop("fecha_texto") # Eliminar la columna de texto original

print("\nDataFrame Corregido:")
print(df_corregido)
print(df_corregido.dtypes)
````

**ExplicaciÃ³n:**
* `.cast(pl.Int64)` convierte la columna "id" a tipo entero.
* `.str.strptime(pl.Date, "%Y-%m-%d")` parsea la cadena de texto "fecha_texto" usando el formato `AÃ±o-Mes-DÃ­a` y la convierte a tipo Fecha (`pl.Date`).
* Usamos `.alias("fecha")` para darle un nuevo nombre y `.drop()` para quitar la original.

Polars ofrece muchos tipos de datos (`pl.Int32`, `pl.Float64`, `pl.Utf8` (texto), `pl.Boolean`, `pl.Datetime`, etc.).

---

## 4. Manejar Datos ErrÃ³neos

Para corregir valores que son lÃ³gicamente incorrectos (pero tienen el tipo correcto), puedes usar:

* **`.filter()`**: Para eliminar filas con valores errÃ³neos.
* **`pl.when().then().otherwise()`**: Para reemplazar valores basados en condiciones.

ğŸ“˜ **Ejemplo: Corregir una duraciÃ³n imposible**

````python
df_erroneo = pl.DataFrame({"duracion_min": [60, 45, 9999, 75, -10]})
print("Datos originales:")
print(df_erroneo)

# OpciÃ³n 1: Eliminar filas con duraciÃ³n > 180 o < 0
df_filtrado_err = df_erroneo.filter(
    (pl.col("duracion_min") <= 180) & (pl.col("duracion_min") >= 0)
)
print("\nFiltrando valores errÃ³neos:")
print(df_filtrado_err)

# OpciÃ³n 2: Reemplazar valores errÃ³neos (ej. con la mediana)
mediana_duracion = df_erroneo.filter(
    (pl.col("duracion_min") <= 180) & (pl.col("duracion_min") >= 0)
)["duracion_min"].median()

df_reemplazado_err = df_erroneo.with_columns([
    pl.when((pl.col("duracion_min") > 180) | (pl.col("duracion_min") < 0))
      .then(mediana_duracion)
      .otherwise(pl.col("duracion_min"))
      .alias("duracion_corregida")
])
print("\nReemplazando valores errÃ³neos con mediana:")
print(df_reemplazado_err)
````

---

## 5. Eliminar Duplicados (`.unique()`)

Para eliminar filas completamente duplicadas, usa `.unique()`.

````python
df_duplicados = pl.DataFrame({
    "a": [1, 2, 1, 3, 2],
    "b": ["x", "y", "x", "z", "y"]
})
print("Con duplicados:")
print(df_duplicados)

df_sin_duplicados = df_duplicados.unique()
print("\nSin duplicados:")
print(df_sin_duplicados)
````

Puedes usar `subset` para considerar duplicados basados solo en ciertas columnas y `keep` para decidir cuÃ¡l fila mantener (`'first'`, `'last'`, `'none'`):
`df_duplicados.unique(subset=["a"], keep='first')`

## âœ… **Resumen de Limpieza**

| Tarea                  | MÃ©todo Polars                             | DescripciÃ³n                                    |

| Rellenar Nulos         | `.fill_null(valor_o_expresiÃ³n)`           | Reemplaza nulos.                               |

| Eliminar Duplicados    | `.unique(subset=..., keep=...)`           | Elimina filas duplicadas.                      |

# **LecciÃ³n: Correlaciones de Datos en Polars** ğŸ“ˆğŸ“‰

---

## 1. CorrelaciÃ³n: Midiendo la RelaciÃ³n entre Variables

La **correlaciÃ³n** es una medida estadÃ­stica que describe la **fuerza y direcciÃ³n de la relaciÃ³n lineal** entre dos variables numÃ©ricas.
Es fundamental en el anÃ¡lisis exploratorio de datos (EDA) y en la selecciÃ³n de caracterÃ­sticas (*feature selection*) para modelos de Machine Learning.

El **coeficiente de correlaciÃ³n** (usualmente Pearson) varÃ­a entre **-1 y +1**:

* **+1**: CorrelaciÃ³n positiva perfecta (si una variable sube, la otra sube proporcionalmente).
* **-1**: CorrelaciÃ³n negativa perfecta (si una variable sube, la otra baja proporcionalmente).
* **0**: No hay correlaciÃ³n lineal (las variables no se mueven juntas de forma lineal).
* Valores cercanos a +1 o -1 indican una relaciÃ³n fuerte; valores cercanos a 0 indican una relaciÃ³n dÃ©bil.

---

## 2. Calcular Correlaciones en Polars

Polars permite calcular la matriz de correlaciÃ³n entre las columnas numÃ©ricas de un DataFrame usando el mÃ©todo `.corr()`.

ğŸ“˜ **Dataset de ejemplo (datos de entrenamiento):**

````python
import polars as pl

url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/training_data.csv'
# Usamos read_csv porque el dataset es pequeÃ±o; para grandes, scan_csv serÃ­a mejor
df_train = pl.read_csv(url)

# Eliminar filas con valores nulos para el cÃ¡lculo de correlaciÃ³n
df_train = df_train.drop_nulls()

print("Datos de entrenamiento (sin nulos):")
print(df_train.head())
````

### ğŸ”¹ Calcular la Matriz de CorrelaciÃ³n

````python
# Calcular la correlaciÃ³n entre todas las columnas numÃ©ricas
corr_matrix = df_train.corr()

print("\nMatriz de CorrelaciÃ³n:")
print(corr_matrix)
````

---

## 3. InterpretaciÃ³n de la Matriz

La tabla resultante muestra el coeficiente de correlaciÃ³n entre cada par de columnas:

* La diagonal siempre es **1.0**, ya que una variable estÃ¡ perfectamente correlacionada consigo misma.
* La matriz es **simÃ©trica**: la correlaciÃ³n entre A y B es la misma que entre B y A.

**Observaciones del ejemplo:**

* `Duration` y `Calories` tienen una **correlaciÃ³n positiva muy alta** (0.92): a mayor duraciÃ³n, mÃ¡s calorÃ­as quemadas (esperado).
* `Pulse` y `Maxpulse` tienen una **correlaciÃ³n positiva fuerte** (0.79): pulsaciones medias y mÃ¡ximas tienden a subir juntas.
* `Duration` y `Maxpulse` tienen una **correlaciÃ³n muy baja** (cercana a 0): no hay relaciÃ³n lineal clara entre cuÃ¡nto dura el ejercicio y cuÃ¡l es el pulso mÃ¡ximo alcanzado en ese ejercicio especÃ­fico.
* `Pulse` y `Calories` tienen una **correlaciÃ³n muy baja** (0.02): el pulso medio por sÃ­ solo no predice bien las calorÃ­as quemadas (la duraciÃ³n es mucho mÃ¡s importante aquÃ­).

---

## 4. VisualizaciÃ³n con un Mapa de Calor (Heatmap)

Una matriz de correlaciÃ³n es mÃ¡s fÃ¡cil de interpretar visualmente usando un mapa de calor. Podemos usar `seaborn` y `matplotlib` (necesitarÃ¡s tenerlos instalados: `pip install seaborn matplotlib`).

**Importante:** Seaborn trabaja directamente con DataFrames de Pandas. Necesitamos convertir la matriz de correlaciÃ³n de Polars a Pandas usando `.to_pandas()`.

````python
import polars as pl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convertir la matriz de correlaciÃ³n de Polars a Pandas para Seaborn
corr_matrix_pd = corr_matrix.to_pandas()
````

````python
print("Matriz convertida a Pandas (lista para heatmap):")
print(corr_matrix_pd)
# ----------------------------------------
print("Forma del DataFrame Pandas:", corr_matrix_pd.shape)
print("Ãndice:", corr_matrix_pd.index)
print("Columnas:", corr_matrix_pd.columns)
print("-" * 30)
# -----------
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix_pd, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de CorrelaciÃ³n de Datos de Entrenamiento')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
````

**InterpretaciÃ³n del Heatmap:**
